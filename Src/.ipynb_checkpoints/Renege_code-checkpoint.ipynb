### now, lets take care of the date columns, we will take the difference between the two and use it as a numeric predictor

```python
# Assuming renege_df is already loaded and cleaned
renege_df['date_1st_contact'] = pd.to_datetime(renege_df['date_1st_contact'], errors='coerce')
renege_df['date_offered'] = pd.to_datetime(renege_df['date_offered'], errors='coerce')

# Calculate the difference in days between date_offered and date_1st_contact
renege_df['days_to_offer'] = (renege_df['date_offered'] - renege_df['date_1st_contact']).dt.days

# Drop the original date columns as they are no longer needed
renege_df.drop(columns=['date_1st_contact', 'date_offered'], inplace=True)

# Check the resulting dataframe
print(renege_df.head())
```

## Checking class inbalance (if dataset is balanced or not)

```python
# Calculate class distribution
class_distribution = renege_df['offer_dropped'].value_counts(normalize=True) * 100

print("Class Distribution:")
print(class_distribution)

# Calculate imbalance metrics
majority_class = class_distribution.idxmax()
minority_class = class_distribution.idxmin()

majority_count = renege_df['offer_dropped'].value_counts().max()
minority_count = renege_df['offer_dropped'].value_counts().min()

ratio = majority_count / minority_count
difference = majority_count - minority_count
difference_percentage = (difference / (majority_count + minority_count)) * 100

print("\nImbalance Metrics:")
print(f"Ratio: {ratio:.2f}")
print(f"Difference: {difference} ({difference_percentage:.2f}%)")


# Determine imbalance
if ratio > 2 or difference_percentage > 10:
    print("\nDataset is imbalanced.")
else:
    print("\nDataset is relatively balanced.")
```

```python
#Creating model for predictors
predictors= ['age', 'marital_status', 'education_level', 'gender', 'percent_hike',
       'distance_from_home', 'sourcing_channel', 'total_rounds',
       'satisfaction_index', 'no_companies_worked', 'career_growth',
       'flexi_work', 'total_experience', 'timely_communication',
        'days_to_offer']
outcome = ['offer_dropped']

# partition data
X = pd.get_dummies(renege_df[predictors], drop_first=True)
y = renege_df[outcome]
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.3, random_state=1) #training data 70% and validation data 30%
```

```python
# Find the best k from 1 to 50
results = []
for k in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors=k).fit(train_X, train_y)
    results.append({
        'k': k,
        'accuracy': accuracy_score(valid_y, knn.predict(valid_X.values))
    })

# Convert results to a pandas data frame
results = pd.DataFrame(results)
print(results)
```

### we can see from the above table that k=5 is giving us the  accuracy of 88 on the validation data

```python
#model with best k, we are moving with k = 5
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(train_X, train_y)
classificationSummary(valid_y, knn_model.predict(valid_X.values))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score

# Predict on validation set
y_pred = knn_model.predict(valid_X.values)

# Evaluate model
print("Accuracy:", accuracy_score(valid_y, y_pred))
print("Precision:", precision_score(valid_y, y_pred))
print("Recall:", recall_score(valid_y, y_pred))
print("F1-score:", f1_score(valid_y, y_pred))
print("Classification Report:")
print(classification_report(valid_y, y_pred))
```

## Now, lets repeat the process by doing the dimensionality reduction using PCA and K-fold and see the new model accuracy

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Apply PCA for dimensionality reduction
pca = PCA(n_components=0.95) 
train_X_pca = pca.fit_transform(train_X)
valid_X_pca = pca.transform(valid_X)

# Define KNN model
knn_model = KNeighborsClassifier(n_neighbors=5)

# Cross-validation with KFold
kf = KFold(n_splits=5, shuffle=True, random_state=1)
cv_scores = cross_val_score(knn_model, train_X_pca, train_y.values.ravel(), cv=kf, scoring='accuracy')
print("Cross-validation scores:", cv_scores)
print("Average CV accuracy:", cv_scores.mean())

# Train KNN model on reduced data
knn_model.fit(train_X_pca, train_y.values.ravel())

# Evaluate model on validation set
y_pred = knn_model.predict(valid_X_pca)
print("Validation accuracy:", accuracy_score(valid_y, y_pred))
print("Validation classification report:")
print(classification_report(valid_y, y_pred))
print("Validation confusion matrix:")
print(confusion_matrix(valid_y, y_pred))


# Hyperparameter tuning
param_grid = {'n_neighbors': [3, 5, 7, 9]}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=kf)
grid_search.fit(train_X_pca, train_y.values.ravel())
print("Best parameters:", grid_search.best_params_)
print("Best accuracy:", grid_search.best_score_)
```

### Observation: I am seeing that by doing PCA and K-fold, I am getting the best accuracy of 86.34% on K=9. Accuracy did not improve much. Earlier it was 85.6%

## using all predictors to create the classification tree, coding the text categories into numbers

```python
from sklearn.preprocessing import OrdinalEncoder

# Extracting all columns except 'Competitive?'
numerical_cols = ['age', 'percent_hike', 'total_rounds', 'satisfaction_index', 'no_companies_worked', 'total_experience', 'days_to_offer']

categorical_cols = ['marital_status', 'education_level', 'gender', 'distance_from_home', 'sourcing_channel', 'career_growth', 'flexi_work', 'timely_communication']
predictor_cols = numerical_cols + categorical_cols  # Define the order of columns for the final DataFrame

# Copying the dataframe with only the categorical columns
X_categorical = renege_df[categorical_cols].copy()

# Initialize the ordinal encoder
ordinal_encoder = OrdinalEncoder()

# Transform the categorical columns
X_encoded = ordinal_encoder.fit_transform(X_categorical)

# Create a DataFrame with encoded categorical columns and selected numerical columns
df_numerical = renege_df[numerical_cols]
df_final = pd.concat([df_numerical, pd.DataFrame(X_encoded, columns=categorical_cols)], axis=1)

# Add 'offer_dropped' column to the end
df_final['offer_dropped'] = renege_df['offer_dropped']

# Rearrange columns to place 'Competitive?' at the end
df_final = df_final[predictor_cols + ['offer_dropped']]

df_final.head(5)
```

## Seeing how the categories are coded

```python
# Access the categories and their respective encoded values
encoded_categories = ordinal_encoder.categories_
for i, col in enumerate(categorical_cols):
    print(f"Column '{col}' categories and encoded values:")
    for j, category in enumerate(encoded_categories[i]):
        print(f"Category '{category}' is encoded as {j}")
    print()
```

## Hyperparameter tuning using GridSearchCV : now, lets use grid search to improve the accuracy further

```python
#Use grid search for the best parameters
# Start with an initial guess for parameters
param_grid = {
    'max_depth': [10, 20, 30, 40], 
    'min_samples_split': [20, 40, 60, 80, 100], 
    'min_impurity_decrease': [0, 0.0005, 0.001, 0.005, 0.01], 
}
gridSearch = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, n_jobs=-1) #n_jobs=-1 means 
        #that the availalbe computer memory (CPU) will be used to make calculations faster. 
gridSearch.fit(train_X, train_y)
print('Initial score: ', gridSearch.best_score_)
print('Initial parameters: ', gridSearch.best_params_)
```

```python

# Adapt grid based on result from initial grid search *separate from code above
param_grid = {
    'max_depth': list(range(5,15)), 
    'min_samples_split': list(range(12, 22)), #keeping splits between 12 to 22 
    'min_impurity_decrease': [0.0009, 0.001, 0.0011], 
}
gridSearch = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, n_jobs=-1)
gridSearch.fit(train_X, train_y)
print('Improved score: ', gridSearch.best_score_)
print('Improved parameters: ', gridSearch.best_params_)

bestClassTree = gridSearch.best_estimator_
```

```python
## Import Necessary Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt


## Preprocess Data
# Define predictors and outcome
predictors = ['age', 'percent_hike', 'total_rounds', 'satisfaction_index', 
              'no_companies_worked', 'total_experience', 'days_to_offer']
categorical_features = ['marital_status', 'education_level', 'gender', 
                        'distance_from_home', 'sourcing_channel', 'career_growth', 
                        'flexi_work', 'timely_communication']
outcome = 'offer_dropped'

# One-Hot Encoding (OHE)
ohe = OneHotEncoder(sparse=False)
ohe_df = ohe.fit_transform(renege_df[categorical_features])
ohe_df = pd.DataFrame(ohe_df, columns=ohe.get_feature_names_out(categorical_features))
renege_df = pd.concat([renege_df, ohe_df], axis=1)

# Convert offer_dropped to binary (0/1)
renege_df[outcome] = renege_df[outcome].cat.codes

# Drop original categorical columns
renege_df.drop(categorical_features, axis=1, inplace=True)


## Split Data
train_df, valid_df = train_test_split(renege_df, test_size=0.2, random_state=42)


## Scale Data
scaler = StandardScaler()
train_df[predictors] = scaler.fit_transform(train_df[predictors])
valid_df[predictors] = scaler.transform(valid_df[predictors])


## Logistic Regression Model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(train_df[predictors + list(ohe_df.columns)], train_df[outcome])

## Make Predictions
train_pred = log_reg.predict(train_df[predictors + list(ohe_df.columns)])
valid_pred = log_reg.predict(valid_df[predictors + list(ohe_df.columns)])

## Evaluate Model
print("Training Accuracy:", accuracy_score(train_df[outcome], train_pred))
print("Validation Accuracy:", accuracy_score(valid_df[outcome], valid_pred))
print("Classification Report:")
print(classification_report(valid_df[outcome], valid_pred))
print("Confusion Matrix:")
print(confusion_matrix(valid_df[outcome], valid_pred))


## Visualize ROC Curve
log_reg_prob = log_reg.predict_proba(valid_df[predictors + list(ohe_df.columns)])[:,1]
fpr, tpr, thresholds = roc_curve(valid_df[outcome], log_reg_prob)
plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```

## Naive logistic regression model is giving us an overall accuracy of 96%

```python
from sklearn.inspection import partial_dependence, PartialDependenceDisplay
import numpy as np
import pandas as pd

# Create a grid of values for 'flexi_work_Yes'
grid_values = np.linspace(0, 1, 10)

# Create a new DataFrame with the same length as grid_values
grid_df = pd.DataFrame(np.repeat(train_df.drop('offer_dropped', axis=1).mean().values, 10).reshape(10, len(train_df.drop('offer_dropped', axis=1).columns)), 
                        columns=train_df.drop('offer_dropped', axis=1).columns)

# Set 'flexi_work_Yes' to the grid values
grid_df['flexi_work_Yes'] = grid_values

# Predict probabilities
preds = log_reg.predict_proba(grid_df)[:, 1]

# Plot partial dependence
plt.plot(grid_values, preds)
plt.xlabel('flexi_work_Yes')
plt.ylabel('Predicted Probability of offer_dropped')
plt.show()
```

## Calculating the net impact of flexi-work on offer dropout rates 

```python
# Predict dropout rates for flexi-work 0 and 1
dropout_rate_0 = log_reg.predict_proba(grid_df[grid_df['flexi_work_Yes'] == 0])[:, 1][0]
dropout_rate_1 = log_reg.predict_proba(grid_df[grid_df['flexi_work_Yes'] == 1])[:, 1][0]

# Calculate absolute difference
difference = abs(dropout_rate_0 - dropout_rate_1)

# Express as percentage
percentage_reduction = (difference / dropout_rate_0) * 100

print(f"Net impact: {percentage_reduction:.2f}% reduction in offer dropout rate")
```
